```python
# Create a new DataFrame to store indicators
indicators = pd.DataFrame()

# Loop over all columns in the original DataFrame
for col in df.columns:
    if df[col].isna().any():

        # Create boolean mask for NaNs in the current column
        indicators[col] = df[col].isna()

        # Rename the boolean column to indicate that it is an indicator
        indicators = indicators.rename(columns={col: col + '_indicator'})

        # Add boolean indicator column to the original DataFrame
        df = pd.concat([df, indicators[col + '_indicator']], axis=1)

```

```python
def plot_these_line_graphs(x_row, y_col):
    x = df[x_row]
    y = df[y_col]
    plt.scatter(x, y, color='red', s=100)
    plt.title('Graph contrasting x and y')
    plt.show()
```
